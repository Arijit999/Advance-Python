```python
'''
Python uses a mechanism called integer interning. 
For small integers (typically from -5 to 256), Python 
reuses the same object to save memory and improve performance.
'''
```

```python
a = 256
b = 256
print(id(a) == id(b))
print(f"{memoryview(bytes(a))}\n{memoryview(bytes(b))}")
```

```python
a = 10000
b = 10000
print(id(a)==id(b))
print(hex(id(a)))
print(hex(id(b)))
```

```python
a = -5
b = -5
print(id(a)==id(b))
print(hex(id(a)))
print(hex(id(b)))
```

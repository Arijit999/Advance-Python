```python
def smart_div(fun):
    def inner (a,b):
        if a<b:
            a,b=b,a
            return fun(a,b)
    return inner
```

```python
def div(a : int , b : int) -> float:
    return a/b

print(div(2,4))
div=smart_div(div)
print(div(2,4))
```

```python
@smart_div
def div (a : int , b : int ) -> float:
    return a/b

print(div(2,4))
```

```python
def div (a : int , b : int ) -> float:
    return( a/b)

print(smart_div(div)(2,4))
```

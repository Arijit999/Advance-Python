##  Decorators

```python
''' 
Decorators changes or modifies 
the behaviour of a function execution
but does not changes the internal 
code of a function.
'''
def smart_div(fun):
    def wrapper (a,b):
        if a<b:
            a,b=b,a
            return fun(a,b)
    return wrapper
```

### 3 ways you can use decorators 

```python
# Creating an instance of decorator
def div(a : int , b : int) -> float:
    return a/b

print(div(2,4)) # Before using decorator 0.5

div=smart_div(div)  

print(div(2,4)) # After using decorator 2.0
```

```python
# Using @
@smart_div 
def div (a : int , b : int ) -> float:
    return a/b

print(div(2,4))
```

```python
# Direct call
def div (a : int , b : int ) -> float:
    return( a/b)

print(smart_div(div)(2,4))
```

```python
 # Example 2
def msg (fun):
    def inner (a,b):
        print("The sum is : ",end="")
        return fun(a,b)
    return inner

@msg
def Sum(a,b):
    return a+b

print(Sum(5,6))

```

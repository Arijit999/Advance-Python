Generators generates values on the fly and does not stores value in memory unlike lists or any iterables.

```python
def mygen():
    for i in range(5):
        yield i
```

```python
gen=mygen()
for i in gen:
    print(i)
```

```python
gen2=mygen()
print(next(gen2))
print(next(gen2))
print(next(gen2))
print(next(gen2))
```
